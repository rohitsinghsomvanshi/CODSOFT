```python
from sklearn.ensemble import RandomForestClassifier # importing the RandomForestClassifier from sklearn.ensemble module. This is a machine learning algorithm used for classification tasks.    
from sklearn.linear_model import LogisticRegression # importing the LogisticRegression from sklearn.linear_model module. This is a machine learning algorithm used for classification tasks.
from sklearn.metrics import (   # importing various metrics from sklearn.metrics module to evaluate the performance of machine learning models.
    accuracy_score,             #( matlb model ki accuracy ko measure karne ke liye use hota hai)
    auc,                           # (matlb area under the curve ko measure karne ke liye use hota hai)
    classification_report,           # (matlb model ke classification report ko generate karne ke liye use hota hai)
    confusion_matrix,               # (matlb model ke confusion matrix ko generate karne ke liye use hota hai)
    f1_score,                       # (matlb model ke f1 score ko measure karne ke liye use hota hai)
    precision_recall_curve,         # (matlb model ke precision-recall curve ko generate karne ke liye use hota hai)    
    precision_score,                # (matlb model ke precision score ko measure karne ke liye use hota hai)
    recall_score,                   # (matlb model ke recall score ko measure karne ke liye use hota hai)
    roc_auc_score,                  # (matlb model ke ROC AUC score ko measure karne ke liye use hota hai)
    roc_curve,                      # (matlb model ke ROC curve ko generate karne ke liye use hota hai) 
)
from sklearn.model_selection import train_test_split  # importing the train_test_split function from sklearn.model_selection module. This function is used to split the dataset into training and testing sets for machine learning models. (matlb dataset ko training aur testing sets me split karne ke liye use hota hai )
from sklearn.preprocessing import StandardScaler       # importing the StandardScaler from sklearn.preprocessing module. This is used to standardize features by removing the mean and scaling to unit variance. (matlb features ko standardize karne ke liye use hota hai) 
```

```python
# Separate features and target.
x = df.drop("Class", axis=1).copy()  # Dropping the "Class" column from the DataFrame to create the feature set (x) for machine learning models. The "Class" column is the target variable that we want to predict, so we remove it from the features. The copy() method is used to create a copy of the DataFrame without the "Class" column, ensuring that the original DataFrame remains unchanged.  (matlb "Class" column ko drop karke feature set (x) create karna, jisme "Class" column target variable hai jise predict karna hai. copy() method ka use karke original DataFrame ko unchanged rakha jata hai.   )
y = df["Class"].copy() # iska matlab hai ki humne "Class" column ko target variable (y) ke roop me copy kiya hai. Ye target variable hai jise hum predict karna chahte hain machine learning models ke liye. (matlb "Class" column ko target variable (y) ke roop me copy karna, jise predict karna hai machine learning models ke liye.)   


print("Target distribution:") 
print(y.value_counts())         # Display the count of each class in the target variable(matlb target variable me har class ka count print karna)
print("\nMissing target values:", y.isnull().sum())
```

```python
# Train-test split before scaling prevents data leakage.
x_train, x_test, y_train, y_test = train_test_split(    # Splitting the dataset into training and testing sets
    x,      # Features (input variables) for the model
    y,      # Target variable (output variable) for the model
    test_size=0.2, # ye data ka 20% test set me jayega aur 80% training set me jayega
    random_state=42, # ye random state ko set karne ke liye use hota hai taki har bar same random split mile
    stratify=y,   # ye stratify parameter ko set karne ke liye use hota hai taki training aur testing sets me target variable ka distribution same rahe
)

scaler = StandardScaler()
scale_cols = ["Amount", "Time"]
x_train.loc[:, scale_cols] = scaler.fit_transform(x_train[scale_cols])
x_test.loc[:, scale_cols] = scaler.transform(x_test[scale_cols])

print("Training set size:", x_train.shape[0])
print("Test set size:", x_test.shape[0])
```

```python
# StandardScaler ka matlab:
# 1. Mean (average) ko zero ke paas karta hai
# 2. Standard deviation (spread) ko 1 karta hai
# Yani values normalize hoti hain taaki ML model better train ho.

print("PEHLE (Scaling se pahle) - Amount column:")
print("Min:", df["Amount"].min())
print("Max:", df["Amount"].max())
print("Mean:", df["Amount"].mean())

print("\nTRAINING DATA (Scaling ke baad) - Amount column:")
print("Min:", x_train["Amount"].min())
print("Max:", x_train["Amount"].max())
print("Mean:", x_train["Amount"].mean())

print("\n" + "=" * 50)
print("Pehle wale 5 Amount values:", df["Amount"].head().values)
print("Scaled training ke 5 Amount values:", x_train["Amount"].head().values)
```

```python
# Logistic Regression with balanced class weights for imbalanced fraud data.
model = LogisticRegression(class_weight="balanced", max_iter=2000, random_state=42)
model.fit(x_train, y_train)

y_prob_model = model.predict_proba(x_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_prob_model)

# F1 values align with precision/recall points. The last PR point has no matching threshold.
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
print("Logistic Regression training complete")
```

```python
# Choose the threshold with the best F1 score from valid threshold points.
if len(thresholds) == 0:
    best_threshold = 0.5
else:
    best_idx = int(np.nanargmax(f1_scores[:-1]))
    best_threshold = float(thresholds[best_idx])

print(f"Best threshold by F1 score: {best_threshold:.6f}")

y_pred_tuned = (y_prob_model >= best_threshold).astype(int)
print(f"\n--- Logistic Regression (tuned threshold: {best_threshold:.6f}) ---\n")
print(classification_report(y_test, y_pred_tuned, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tuned))
print("ROC AUC Score:", roc_auc_score(y_test, y_prob_model))
```

```python
# Random Forest Classifier
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced_subsample",
    n_jobs=1,
)
rf_model.fit(x_train, y_train)

rf_pred = rf_model.predict(x_test)
y_prob_rf = rf_model.predict_proba(x_test)[:, 1]

print("\n--- Random Forest Classifier ---\n")
print(classification_report(y_test, rf_pred, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))
print("Random Forest ROC AUC Score:", roc_auc_score(y_test, y_prob_rf))
```
